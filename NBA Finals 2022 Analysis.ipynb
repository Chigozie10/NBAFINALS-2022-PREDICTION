```python
import matplotlib.pyplot as plt

df

year_group = df1.groupby('Year')

avg_celts = df1.groupby('Year').mean()['Celtics score']

avg_warriors = df1.groupby('Year').mean()['Warriors score']

avg_celts

years = [year for year, df in year_group]

years


# Plot the average point scored by the Celtics over the years against the Warriors

plt.plot(years, avg_celts)
plt.xticks(years, rotation = 'vertical', size = 8)
plt.ylabel('Average Points for Celtics')
plt.xlabel('Years')
plt.title('Average Points by the Celtics against the Warriors over the years')
plt.grid()
plt.show()
```

```python


fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.plot(years, avg_celts, color ='g')
ax2.plot(years, avg_warriors, 'b-')

ax1.set_xlabel('Years')
ax1.set_ylabel('Celtics Points', color ='g')
ax2.set_ylabel('Warriors Points', color ='b')
ax1.set_xticklabels(years, rotation = 'vertical', size = 8)

plt.title('Average Points by the Celtics and Warriors against each other over the years')
plt.grid()
plt.show()

# Plot the average point scored by the Celtics over the years against the Warriors

# plt.plot(years, avg_celts)
# plt.xticks(years, rotation = 'vertical', size = 8)
# plt.ylabel('Average Points for Celtics')
# plt.xlabel('Years')
# plt.title('Average Points by the Celtics against the Warriors over the years')
# plt.grid()
# plt.show()

```
